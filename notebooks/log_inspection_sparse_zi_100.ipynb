```python
import pandas as pd
from pathlib import Path
pd.set_option('display.max_rows', 50)        # how many rows to show at once
pd.set_option('display.max_colwidth', None)  # don’t truncate text columns


run_id = 'sparse_zi_100_log'  # TODO: change to the log directory you want to inspect
log_dir = Path('..') / 'log' / run_id
sorted(log_dir.iterdir())  # preview files
```

```python
summary = pd.read_pickle(log_dir / 'summary_log.bz2')
summary.head(104)
```

```python
fundamental = pd.read_pickle(log_dir / 'fundamental_JPM.bz2')
fundamental = fundamental.reset_index()   # brings FundamentalTime back as a column
fundamental.plot(x='FundamentalTime', y='FundamentalValue', legend=False)
```
